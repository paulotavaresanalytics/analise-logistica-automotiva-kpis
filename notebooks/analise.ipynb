```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/dados_frota.csv')

df['custo_km'] = df['custo_manutencao'] / df['km_rodado']
df['consumo_km_l'] = df['km_rodado'] / df['combustivel_litros']

tempo_medio = df['tempo_entrega_horas'].mean()
print('Tempo médio de entrega:', tempo_medio)

df.plot(x='veiculo', y='custo_km', kind='bar', title='Custo por KM')
plt.show()
```
